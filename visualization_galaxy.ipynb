```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
## 전체 증가율 데이터 셋 불러오기
data = pd.read_csv('analysis_2.csv')
data
```

```python
## 갤럭시 모델만 비교하기 위한 인덱싱
model = data['기종'][:3]
before = data['출시 전'][:3]
after = data['출시 후'][:3]
```

```python
## 시각화

fig, ax = plt.subplots(figsize = (12, 6))
bar_width = 0.3

index = np.arange(3)

bar1 = plt.bar(index, before, bar_width, alpha = 0.4, color = 'pink', label = '출시 전')
bar2 = plt.bar(index + bar_width, after, bar_width, alpha = 0.4, color = 'skyblue', label = '출시 후')

plt.xticks(index + bar_width / 2, model)

plt.title('스마트폰 출시 전 후 증가율 비교')
plt.xlabel('갤럭시 모델')
plt.ylabel('평균 조회수 증가율 (%)')
plt.legend()
plt.show()
```
